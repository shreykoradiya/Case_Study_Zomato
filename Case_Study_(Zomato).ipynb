## Ten Highest number of restaurants

```python
import plotly.express as px
```

```python
res=data['Restaurant Name'].value_counts().head(10)
px.bar(res)
```

```python
rest_dining = data.groupby('Restaurant Name')['Dining Rating'].mean().reset_index()
sorted_dining = rest_dining.sort_values('Dining Rating', ascending=False).head(10)

sorted_dining
```

```python
restaurant = sorted_dining['Restaurant Name']
ratings = sorted_dining['Dining Rating']

plt.pie(ratings, labels = restaurant, hatch=['.', 'o', 'O'])
plt.title('Top 10 restaurants by dining rating')
plt.show()
```

```python
rest_delivery = data.groupby('Restaurant Name')['Delivery Rating'].mean().reset_index()
sorted_delivery = rest_delivery.sort_values('Delivery Rating', ascending=False).head(10)
sorted_delivery
```

```python
rest_name = sorted_delivery['Restaurant Name']
delivery_rating = sorted_delivery['Delivery Rating']
colors = ['#FFA07A', '#FFB38D', '#FFCBA4', '#FFE1BD', '#FFFFCC', '#D4EFDF', '#AED6F1', '#F8C471', '#D2B4DE', '#F9E79F']

_,_,autotexts=plt.pie(delivery_rating, labels=rest_name, colors= colors, autopct='')

for i, delivery_rating in enumerate(delivery_rating):
    autotexts[i].set_text(f'{delivery_rating:.1f}')
```

```python
city_wise_best_dishes = data.loc[data.groupby('City')['Votes'].idxmax()][
    ['City', 'Item Name', 'Votes', 'Restaurant Name']
].reset_index(drop=True)

city_wise_best_dishes.sort_values('City', inplace=True)

city_wise_best_dishes
```

```python
best_restaurant = data.loc[data.groupby('Item Name')['Votes'].idxmax()][
    ['Item Name', 'Restaurant Name']
].reset_index(drop=True)

best_restaurant.sort_values('Item Name', inplace=True)

best_restaurant
```

```python
best_restaurant_city = data.loc[data.groupby('Item Name')['Votes'].idxmax()][
    ['Item Name', 'Restaurant Name', 'City']
].reset_index(drop=True)

best_restaurant_city.sort_values('Item Name', inplace=True)

best_restaurant_city
```
